```python
import requests
import time
import json
import csv
import numpy as np
import pandas as pd
from google.colab import userdata
```

```python
# Set up your GitHub API token here (optional, but recommended for higher rate limits)
GITHUB_TOKEN = userdata.get('GIT_TOKEN')
headers = {"Authorization": f"token {GITHUB_TOKEN}"}
```

```python
# Define query parameters
location = "Chennai"
min_followers = 50
search_url = f"https://api.github.com/search/users?q=location:{location}+followers:>{min_followers}"

# Function to get users based on search criteria
def get_users_from_chennai(search_url):
    users = []
    page = 1
    while True:
        response = requests.get(search_url + f"&page={page}", headers=headers)
        if response.status_code != 200:
            print("Error:", response.json())
            break

        data = response.json()
        users.extend(data['items'])

        # Check if there are more pages
        if "next" in response.links:
            page += 1
        else:
            break

    return users

# Function to fetch additional details for each user
def get_user_details(username):
    user_url = f"https://api.github.com/users/{username}"
    response = requests.get(user_url, headers=headers)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Failed to fetch details for {username}: {response.status_code}")
        return None

# Get initial list of users
users = get_users_from_chennai(search_url)

# Retrieve additional details for each user and save to CSV
with open("users.csv", mode="w", newline='', encoding="utf-8") as file:
    writer = csv.writer(file)
    # Write the header
    writer.writerow(["login", "Name", "company", "location", "email", "hireable",
                     "bio", "public_repos", "followers", "following", "created_at"])

    for user in users:
        user_details = get_user_details(user['login'])
        if user_details:
            writer.writerow([
                user_details['login'],
                user_details.get('name'),
                user_details.get('company'),
                user_details.get('location'),
                user_details.get('email'),
                user_details.get('hireable'),
                user_details.get('bio'),
                user_details.get('public_repos'),
                user_details.get('followers'),
                user_details.get('following'),
                #user_details.get('blog'),
                user_details.get('created_at'),
                #user_details.get('html_url')
            ])
            print(user_details.get('html_url'))
        time.sleep(5)  # Adding delay to avoid hitting rate limits

print("User data saved to chennai_github_users.csv")
```

```python
# Function to fetch repositories for a user
def get_user_repositories(username, max_repos=500):
    repos_url = f"https://api.github.com/users/{username}/repos"
    params = {
        "sort": "pushed",  # Sort by the most recently pushed repositories
        "per_page": 100    # Get up to 100 results per page
    }
    repos = []
    page = 1

    while len(repos) < max_repos:
        response = requests.get(repos_url, headers=headers, params={**params, "page": page})
        if response.status_code != 200:
            print(f"Failed to fetch repositories for {username}: {response.status_code}")
            break

        data = response.json()
        repos.extend(data)

        # Stop if there are no more results
        if len(data) < 100:
            break
        page += 1
        time.sleep(5)  # Avoid rate limits

    return repos[:max_repos]

# Open the users CSV file
with open("users.csv", mode="r", newline='', encoding="utf-8") as users_file, \
     open("user_repos_test.csv", mode="w", newline='', encoding="utf-8") as repos_file:

    user_reader = csv.DictReader(users_file)
    repo_writer = csv.writer(repos_file)

    # Write header for the repositories CSV file
    repo_writer.writerow([
        "login", "full_name", "created_at", "stargazers_count", "watchers_count",
        "language", "has_projects", "has_wiki", "license_name"
    ])

    # For each user, fetch their repositories and write details to CSV
    for user in user_reader:
        username = user['login']
        print(f"Fetching repositories for {username}...")

        repositories = get_user_repositories(username)

        for repo in repositories:
            print(json.dumps(repo, indent=4))  # Pretty print the JSON data)
            repo_writer.writerow([
                username,                        # User login
                repo['full_name'],               # Full name of the repository
                repo['created_at'],              # Creation date of the repository
                repo['stargazers_count'],        # Number of stars
                repo['watchers_count'],          # Number of watchers
                repo['language'],                # Language of the repository
                repo['has_projects'],            # Has projects enabled
                repo['has_wiki'],                # Has wiki enabled
                repo['license']['name'] if repo['license'] else None  # License name (if any)
            ])
        time.sleep(2)  # Delay to respect API rate limits
        break
print("Repository data saved to user_repos.csv")
```

```python
users_df = pd.read_csv('users.csv')
repos_df = pd.read_csv('repositories.csv')
users_df.head()
```

```python
def clean_company_name(company):
    """
    Cleans up company names:
    - Trims whitespace
    - Strips leading "@" (only the first one)
    - Converts to UPPERCASE
    """
    if pd.isnull(company):  # Handle missing values
        return None
    company = company.strip()
    if company.startswith("@"):
        company = company[1:]
    return company.upper()

# Apply the cleaning function to the 'company' column
users_df['company'] = users_df['company'].apply(clean_company_name)
```

```python
import statsmodels.formula.api as sm

# Filter out users without bios
users_df_with_bios = users_df[users_df['bio'].notna()]

# Calculate bio length (word count)
users_df_with_bios['bio_length'] = users_df_with_bios['bio'].apply(lambda bio: len(bio.split()))

# Define the regression model
model = sm.ols(formula='followers ~ bio_length', data=users_df_with_bios)

# Fit the model
results = model.fit()

# Get the regression slope (coefficient for bio_length)
slope = results.params['bio_length']

# Print the slope rounded to 3 decimal places
print(f"{slope:.3f}")
```

```python
import pandas as pd

# Convert 'created_at' to datetime objects with UTC timezone
repos_df['created_at'] = pd.to_datetime(repos_df['created_at'], utc=True)

# Filter for repositories created on weekends (Saturday or Sunday)
weekend_repos = repos_df[repos_df['created_at'].dt.dayofweek.isin([5, 6])]

# Group by 'login' and count the number of repositories created
repo_counts = weekend_repos.groupby('login')['created_at'].count()

# Sort in descending order and get the top 5 users
top_5_users = repo_counts.sort_values(ascending=False).head(5).index.tolist()

# Print the top 5 users' logins, comma-separated
print(','.join(top_5_users))
```

```python
import pandas as pd

# Filter out users with missing names
users_with_names = users_df[users_df['name'].notna()]

# Extract surnames (last word in name)
users_with_names['surname'] = users_with_names['name'].str.strip().str.split().str[-1]

# Count surname occurrences
surname_counts = users_with_names['surname'].value_counts()

# Find the most common surname(s)
max_count = surname_counts.max()
most_common_surnames = surname_counts[surname_counts == max_count].index.tolist()

# Sort alphabetically and print comma-separated
most_common_surnames.sort()
print(','.join(most_common_surnames))
```
